# Example Use

Following is an example use of the library defined above. You can see that we create the simulation object, define any custom variables, tell it to configure the kernels, and in this case we do a single-run execution instead of running a monte carlo. 

```python
import matplotlib.pyplot as plt
import agnus
```

```python
sim = agnus.Simulation()
sim.define_simulation(agents=20000,inf_range=10, inf_time=100, inf_chance=0.3)
sim.configure_kernels()
data = sim.run_once(500)
```

```python
population = {"uninfected": data[0], 
              "infected": data[1], 
              "recovered": data[2]}

fig, ax = plt.subplots()
ax.stackplot(list(range(len(data[0]))), population.values(), labels=population.keys())
ax.legend(loc='upper left')
ax.set_title("Infection Spread")
ax.set_ylabel("Population")
ax.set_xlabel("Epoch")
x, y = 6.4, 4.8
fig.set_size_inches(x*2, y*2)
```
